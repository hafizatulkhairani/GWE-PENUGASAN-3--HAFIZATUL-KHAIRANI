## BAB IV: Univariate Analysis (Visualizing Distributions)

### Soal 8: Distribusi Data Kategori
**Deskripsi:** Buat visualisasi Bar Chart untuk melihat sebaran frekuensi jumlah transaksi pada setiap `Category`.
**Hint:** Gunakan `df['Category'].value_counts().plot(kind='bar')` atau `sns.countplot()`.

```python
# Tulis kode untuk Soal 8 di sini
plt.Figure(figsize=(10,6))
df['Category'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Jumlah Transaksi per Kategori Produk')
plt.xlabel('Kategori')
plt.ylabel('Jumlah Transaksi')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Tuliskan Analisis Output Soal 8:**
- Bar chart menunjukkan distribusi jumlah transaksi di setiap kategori produk. Kategori tertelectronics mendominasi dengan jumlah transaksi lebih tinggi. Ini berguna untuk memahami bahwa electronics paling sering dijual

### Soal 9: Distribusi Data Numerik
